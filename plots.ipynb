```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_c1 = pd.read_csv("./cmake-build-debug/cone1.csv")
df_x1 = pd.read_csv("./cmake-build-debug/xyz1.csv")
```

```python
columns = ['phi', 'z','omega','vz', 'E', 'E_anal']
names = [r'$\varphi [rad]$', r'$z [m]$',r'$\omega [rad/s]$', r'$v_{z} [m/s]$', 'E [J]']
colors = ['steelblue', 'indigo', 'firebrick']
```

```python
for k in range (0,5):
    plt.figure(figsize=(5, 4))  
    if k==4:
        plt.plot(df_c1['t'], df_c1[columns[k]], label=r"$E_{num}$", color=colors[0])
        plt.plot(df_c1['t'], df_c1[columns[k+1]], label=r"$E_{anal}$", color=colors[2])
    else:
        plt.plot(df_c1['t'], df_c1[columns[k]], color=colors[0])
   
    plt.xlabel("t [s]")
    plt.ylabel(names[k])
    #plt.xlim(0, 10)
    plt.grid(True)
    if(k==4):
        plt.legend(loc="lower right")
    plt.tight_layout()
    plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

df_x1 = pd.read_csv("./cmake-build-debug/xyz1.csv")

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.plot(df_x1['y'], df_x1['x'], df_x1['z'], color='steelblue', linewidth=1)

ax.set_xlabel("y [m]")
ax.set_ylabel("x [m]")
ax.set_zlabel("z [m]", labelpad=0.5)  

ax.view_init(elev=25, azim=50)
plt.show()
```
