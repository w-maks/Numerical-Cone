```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_c1 = pd.read_csv("./cmake-build-debug/cone1.csv")
df_x1 = pd.read_csv("./cmake-build-debug/xyz1.csv")
```

```python
df_c2 = pd.read_csv("./cmake-build-debug/cone2.csv")
df_x2 = pd.read_csv("./cmake-build-debug/xyz2.csv")
```

```python
dfs = {r"$\varphi_0 = 1.1, z_0=1, \omega_0=0, v_{z0}=0$": df_c1, r"$\varphi_0 = 0.1, z_0=0.5, \omega_0=2, v_{z0}=1$": df_c2}
columns = ['phi', 'z','omega','vz', 'E', 'E_anal']
names = [r'$\varphi [rad]$', r'$z [m]$',r'$\omega [rad/s]$', r'$v_{z} [m/s]$']
colors = ['indigo', 'steelblue', 'firebrick', 'palevioletred']
```

```python
i=0
for k in range (0,4):
    plt.figure(figsize=(5, 4))  
    for wp, df in dfs.items():
        plt.plot(df['t'], df[columns[k]], color=colors[i], label=wp)
        i+=1
    i=0
    plt.xlabel("t [s]")
    plt.ylabel(names[k])
    plt.xlim(0,50)
    plt.grid(True)
    plt.legend(loc="lower right")
    plt.tight_layout()
    plt.show()
```

```python
for wp, df in dfs.items():
    plt.figure(figsize=(5, 4))  
    plt.plot(df['t'], df['E'], label=r"$E_{num}$", color=colors[i])
    plt.plot(df['t'], df['E_anal'], label=r"$E_{anal}$", color=colors[i+2])
    i+=1
    i=0
    plt.xlabel("t [s]")
    plt.ylabel('E[J]')
    plt.xlim(0,50)
    plt.grid(True)
    plt.legend(loc="lower right")
    plt.tight_layout()
    plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.plot(df_x1['y'], df_x1['x'], df_x1['z'], color='indigo', linewidth=1)

ax.set_xlabel("y [m]")
ax.set_ylabel("x [m]")
ax.set_zlabel("z [m]", labelpad=0.5)  

ax.view_init(elev=25, azim=50)
plt.show()
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.plot(df_x2['y'], df_x2['x'], df_x2['z'], color='steelblue', linewidth=1)

ax.set_xlabel("y [m]")
ax.set_ylabel("x [m]")
ax.set_zlabel("z [m]", labelpad=0.5)  

ax.view_init(elev=25, azim=50)
plt.show()
```
