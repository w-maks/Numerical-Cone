```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_c1 = pd.read_csv("./cmake-build-debug/cone1.csv")
df_x1 = pd.read_csv("./cmake-build-debug/xyz1.csv")
```

```python
df_c2 = pd.read_csv("./cmake-build-debug/cone2.csv")
df_x2 = pd.read_csv("./cmake-build-debug/xyz2.csv")
```

```python
dfs = {r"$\varphi_0 = 1.1, z_0=1, \omega_0=0, v_{z0}=0$": df_c1, r"$\varphi_0 = \pi/0.99, z_0=10, \omega_0=0, v_{z0}=0$": df_c2}
columns = ['phi','z','omega','vz', 'E', 'E_anal']
names = [r'$\varphi [rad]$', r'$z [m]$',r'$\omega [rad/s]$', r'$v_{z} [m/s]$']
colors = ['indigo', 'steelblue', 'firebrick', 'palevioletred']
```

```python
wp_keys = list(dfs.keys())
df1 = dfs[wp_keys[0]]
df2 = dfs[wp_keys[1]]

for k in range(4): 
    fig, axs = plt.subplots(1, 2, figsize=(10, 4))  

    axs[0].plot(df1['t'], df1[columns[k]], color=colors[0], label=wp_keys[0])
    axs[0].set_xlabel("t [s]")
    axs[0].set_ylabel(names[k])
    axs[0].set_xlim(0, 50)
    axs[0].grid(True)
    axs[0].legend(loc="best")

    axs[1].plot(df2['t'], df2[columns[k]], color=colors[1], label=wp_keys[1])
    axs[1].set_xlabel("t [s]")
    axs[1].set_ylabel(names[k])
    axs[1].set_xlim(0, 50)
    axs[1].grid(True)
    axs[1].legend(loc="best")

    plt.tight_layout()
    plt.show()
```
